# Sales Data Analysis Project

## Objective
This project analyzes a retail dataset to:
- Understand customer spending behavior
- Identify trends in sales
- Visualize key insights using graphs

## Tools Used
- Python
- Pandas
- Matplotlib
- Seaborn

## Dataset
We are using a sample dataset that contains:
- Total bill amount
- Tip amount
- Gender
- Day and time of transaction


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style setup
sns.set_style("whitegrid")

print("Libraries imported successfully!")
```

```python
# Load dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

# Show first 5 rows
df.head()
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=df)
```

## Data Exploration
We will inspect the dataset structure, check for missing values, and summarize statistics.


```python
# Dataset info
df.info()

# Statistical summary
df.describe()

# Check missing values
df.isnull().sum()
```

## Basic Data Analysis
We calculate total sales, average tip, and sales grouped by day.


```python
# Total sales
total_sales = df['total_bill'].sum()
print("Total Sales:", total_sales)

# Average tip
avg_tip = df['tip'].mean()
print("Average Tip:", avg_tip)

# Sales per day
sales_per_day = df.groupby('day')['total_bill'].sum()
print("\nSales per Day:\n", sales_per_day)
```

##  Data Visualization
We will create charts to better understand the data.


```python
# Bar chart: Sales per day
sales_per_day.plot(kind='bar')
plt.title("Total Sales per Day")
plt.xlabel("Day")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Scatter plot: Total bill vs Tip
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Tip vs Total Bill")
plt.show()
```

```python
# Histogram: Distribution of total bill
sns.histplot(df['total_bill'], bins=20)
plt.title("Distribution of Total Bills")
plt.show()
```

## Insights

Based on the analysis:
- Customers who spend more tend to give higher tips.
- Sales are higher during weekends (Saturday and Sunday).
- Most transactions fall within a moderate price range.
